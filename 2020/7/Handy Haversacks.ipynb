```python
import re
```

```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
class Node:
    def __init__(self, name):
        self.name = name
        self.children = {}
        self.parents = {}
        
    def __str__(self):
        return ', '.join(self.parents.keys()) + '->' + self.name + '->' + ', '.join(self.children)
    
    def __repr__(self):
        return str(self)

nodes = {}

def parse_line(line):
    parent, children = re.match(r'^(.*) bags contain (.*)\.\n$', line).groups()
    
    if parent in nodes:
        parent_node = nodes[parent]
    else:
        parent_node = Node(parent)
        nodes[parent] = parent_node
        
    if children.strip() == 'no other bags':
        return
    children = [re.match(r'^(\d+) (.*) bag(s*)$', c.strip()).groups() for c in children.split(',')]
    
    for child_groups in children:
        count = int(child_groups[0])
        child = child_groups[1]
        if child in nodes:
            child_node = nodes[child]
        else:
            child_node = Node(child)
            nodes[child] = child_node
            
        child_node.parents[parent] = parent_node
        parent_node.children[child] = (count, child_node)
    

for line in lines:
    parse_line(line)
```

```python
# Part 1
count = 0
visited = {}

def traverse_parents(node):
    global count
    visited[node.name] = True
    count += 1
    for p in node.parents:
        if p not in visited:
            traverse_parents(node.parents[p])
            
traverse_parents(nodes['shiny gold'])
count - 1 # -1 for the shiny gold bag itself
```

```python
# Part 2
def traverse_children(node):
    count = 0
    for child_count, child_node in node.children.values():
        count += child_count + child_count * traverse_children(child_node)
    return count

traverse_children(nodes['shiny gold'])
```
