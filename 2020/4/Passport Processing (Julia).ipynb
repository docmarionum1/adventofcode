```julia
# Part 1

required = ["byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"]
function validate(record)
    for field in required
        if !occursin(field, record)
            return false
        end
    end
    return true
end

num_valid = 0
for record in split(s, "\n\n")
    valid = validate(record)
    # println("$record $valid")
    num_valid += valid
end
num_valid
```

```julia
# Part 2

function validate(record)
    dict = Dict(split(i, ":")[1] => split(i, ":")[2] for i in split(record))
    
    for field in required
        if !haskey(dict, field)
            return false
        end
    end
    
    byr = parse(Int64, dict["byr"])
    if (byr < 1920) | (byr > 2002)
        return false
    end
    
    iyr = parse(Int64, dict["iyr"])
    if (iyr < 2010) | (iyr > 2020)
        return false
    end
    
    eyr = parse(Int64, dict["eyr"])
    if (eyr < 2020) | (eyr > 2030)
        return false
    end
    
    
    m = match(r"^(\d*)(cm|in)$", dict["hgt"])
    if m === nothing
        return false
    end
    
    value = parse(Int64, m.captures[1])
    unit = m.captures[2]
    
    if unit == "cm"
        if (value < 150) | (value > 193)
            return false
        end
    elseif unit == "in"
        if (value < 59) | (value > 76)
            return false
        end
    end
    
    if match(r"^#[0-9a-f]{6}$", dict["hcl"]) === nothing
        return false
    end
    
    if match(r"^amb|blu|brn|gry|grn|hzl|oth$", dict["ecl"]) === nothing
        return false
    end
    
    if match(r"^\d{9}$", dict["pid"]) === nothing
        return false
    end
    
    return true
end

num_valid = 0
for record in split(s, "\n\n")
    valid = validate(record)
    num_valid += valid
end
num_valid
```
