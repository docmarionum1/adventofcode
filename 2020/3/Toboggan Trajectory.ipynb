```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
lines = [l.strip() for l in lines]
lines
```

```python
width = len(lines[0])
height = len(lines)
width, height
```

```python
# Part 1

trees = 0
x = 0
y = 0
dx = 3
dy = 1

while y < height:
    if lines[y][x] == "#":
        trees += 1
        
    x = (x + dx) % width
    y = y + dy
    
trees
```

```python
# Part 2

from functools import reduce
import operator

def f(slope):
    dx, dy = slope
    trees = 0
    x = 0
    y = 0
    
    while y < height:
        if lines[y][x] == "#":
            trees += 1

        x = (x + dx) % width
        y = y + dy

    return trees

reduce(operator.mul, map(f, [(1, 1), (3, 1), (5, 1), (7, 1), (1, 2)]))
```
