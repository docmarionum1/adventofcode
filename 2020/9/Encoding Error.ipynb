```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
numbers = [int(l) for l in lines]
```

```python
# Part 1
def is_valid(array, num):
    for i in range(len(array)):
        for j in array[:i] + array[i+1:]:
            if array[i] + j == num:
                return True
    return False

preamble_len = 25
i = preamble_len
while i < len(numbers):
    num = numbers[i]
    if not is_valid(numbers[i-preamble_len:i], num):
        print(num)
        break
    i += 1
```
