```python
with open('input.txt') as f:
    lines = f.readlines()
```

```python
lines
```

```python
numbers = [int(line) for line in lines]
numbers
```

```python
# Part 1 - Brute force
for i in numbers:
    for j in numbers:
        if i + j == 2020:
            print(i, j, i*j)
```

```python
# Part 2 - Brute force
for i in numbers:
    for j in numbers:
        for k in numbers:
            if i + j + k == 2020:
                print(i, j, k, i*j*k)
```
