```python
import re
```

```python
with open("input.txt") as f:
    lines = f.readlines()
lines
```

```python
# Part 1
lines_1 = [l for l in lines]
accumulator = 0
pointer = 0
while True:
    if lines_1[pointer] == '':
        break
    instruction, argument = re.match(r'^(acc|nop|jmp) ([-|+]\d+)\n$', lines_1[pointer]).groups()
    argument = int(argument)
    
    # Mark the line as visited by clearing it
    lines_1[pointer] = ''
    
    if instruction == 'jmp':
        pointer += argument
    else:
        pointer += 1
        if instruction == 'acc':
            accumulator += argument
        
accumulator    
```

```python
# Part 2 - Brute Force
def test_program(lines):
    accumulator = 0
    pointer = 0
    previous_pointer = -1
    while True:
        if pointer >= len(lines):
            print(pointer, len(lines), accumulator)
        if lines[pointer] == '':
            break
        instruction, argument = re.match(r'^(acc|nop|jmp) ([-|+]\d+)\n$', lines[pointer]).groups()
        argument = int(argument)

        # Mark the line as visited by clearing it
        lines[pointer] = ''

        previous_pointer = pointer

        if instruction == 'jmp':
            pointer += argument
        else:
            pointer += 1
            if instruction == 'acc':
                accumulator += argument
        
    print(accumulator, previous_pointer, pointer, lines[previous_pointer])
    
for i in range(len(lines)):
    print(i, lines[i])
    lines_copy = None
    if 'nop' in lines[i]:
        lines_copy = [l for l in lines]
        lines_copy[i] = lines[i].replace('nop', 'jmp')
    elif 'jmp' in lines[i]:
        lines_copy = [l for l in lines]
        lines_copy[i] = lines[i].replace('jmp', 'nop')
    
    if lines_copy:
        test_program(lines_copy)
```
