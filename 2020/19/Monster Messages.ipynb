```python
import re
```

```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
rules = {}
messages = []

i = 0
while lines[i] != '\n':
    line = lines[i]
    num, rule = re.match(r'^(\d+): (.*)\n$', line).groups()
    rules[num] = rule

    i += 1
    
messages = lines[i+1:]
```

```python
# Part 1

def expand_rule(num):
    rule = rules[num]
    parts = []
    for part in rule.split('|'):
        regex = '('
        
        for subpart in part.strip().split(' '):
            if re.match(r'\d+', subpart):
                regex += expand_rule(subpart)
            else:
                regex += subpart.replace('"', '')
                
        regex += ')'
        
        if len(regex) == 3:
            regex = regex[1]
        
        parts.append(regex)
        
    return '(' + '|'.join(parts) + ')'
        
regex = r'^' + expand_rule('0') + '$'
len([m for m in messages if re.match(regex, m)])
```

```python
# Part 2

rules['8'] = '42 | 42 8'
rules['11'] = '42 31 | 42 11 31'

def expand_rule(num):
    if num == '8':
        return '(' + expand_rule('42') + '+)'
    elif num == '11':
        # Hard code the options of 1-10 instances of 42 31
        # For example, 42 31 | 42 42 42 31 31 31
        return '(' + "|".join(f"(({expand_rule('42')}){{{i}}}({expand_rule('31')}){{{i}}})" for i in range(1, 10)) + ')'
    rule = rules[num]
    parts = []
    for part in rule.split('|'):
        regex = '('
        
        for subpart in part.strip().split(' '):
            if re.match(r'\d+', subpart):
                regex += expand_rule(subpart)
            else:
                regex += subpart.replace('"', '')
                
        regex += ')'
        
        if len(regex) == 3:
            regex = regex[1]
        
        parts.append(regex)
        
    return '(' + '|'.join(parts) + ')'
        
regex = r'^' + expand_rule('0') + '$'
len([m for m in messages if re.match(regex, m)])
```
