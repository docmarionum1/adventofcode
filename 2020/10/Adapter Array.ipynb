```python
import pandas as pd
```

```python
df = pd.read_csv('input.txt', header=None)
df = pd.concat([df, pd.Series([0, df[0].max()+3])]).sort_values(0)
df
```

```python
# Part 1
(df[0].diff() == 1).sum() * ((df[0].diff() == 3).sum())
```

```python
# Part 2
memo = {}
array = df[0].values
def count(i):
    if i in memo:
        return memo[i]
    
    r = 0
    j = i + 1
    
    if j >= len(array):
        return 1
    
    while (j < len(array)) and (array[j] - array[i] < 4):
        r += count(j)
        j += 1
        
    memo[i] = r
    return r

count(0)
```
