```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
# Part 1

def validate_password(line):
    policy, password = [part.strip() for part in line.split(':')]
    count, letter = policy.split(' ')
    min_count, max_count = [int(num) for num in count.split('-')]
    
    num_letters = len([c for c in password if c == letter])
    
    return num_letters >= min_count and num_letters <= max_count

sum(map(validate_password, lines))
```

```python
# Part 2

def validate_password(line):
    policy, password = [part.strip() for part in line.split(':')]
    positions, letter = policy.split(' ')
    position_i, position_j = [int(num) - 1 for num in positions.split('-')]
    
    return (password[position_i] == letter and password[position_j] != letter) or (password[position_j] == letter and password[position_i] != letter)

sum(map(validate_password, lines))
```
