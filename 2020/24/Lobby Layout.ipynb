```python
import numpy as np
```

```python
with open('input.txt') as f:
    lines = [l.strip() for l in f.readlines()]
lines
```

```python
# Part 1
grid = np.zeros((400,400), int)

for line in lines:
    pointer = (200,200)
    i = 0
    while i < len(line):
        #print(pointer)
        x = y = 0
        
        if line[i] == 'e':
            x = 2
        elif line[i] == 'w':
            x = -2
        else:
            if line[i] == 's':
                y = -1
            else:
                y = 1
                
            i += 1
            
            if line[i] == 'e':
                x = 1
            else:
                x = -1
                
        pointer = (pointer[0] + x, pointer[1] + y)
        i += 1
    
    #print(pointer, grid[pointer])
    if grid[pointer] == 0:
        grid[pointer] = 1
    else:
        grid[pointer] = 0
    
grid.sum()
```

```python
# Part 2

for day in range(100):
    new_grid = grid.copy()
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            # Count neighbors
            num_black = 0
            if (i-2 >= 0):
                num_black += grid[i-2,j]
            if (i+2 < grid.shape[0]):
                num_black += grid[i+2,j]
            if (i+1 < grid.shape[0]) and (j+1 < grid.shape[1]):
                num_black += grid[i+1,j+1]
            if (i-1 >= 0) and (j-1 >= 0):
                num_black += grid[i-1,j-1]
            if (i+1 < grid.shape[0]) and (j-1 >= 0):
                num_black += grid[i+1,j-1]
            if (i-1 >= 0) and (j+1 < grid.shape[1]):
                num_black += grid[i-1,j+1]
                
            if (grid[i,j] == 0) and (num_black == 2):
                new_grid[i,j] = 1
            elif (grid[i,j] == 1) and ((num_black == 0) or (num_black > 2)):
                new_grid[i,j] = 0
                
    grid = new_grid
    
new_grid.sum()
```
