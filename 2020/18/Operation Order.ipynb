```python
import re
from functools import reduce  # Required in Python 3
import operator
```

```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
# Part 1

def eval_paren(exp, i):
    count = 1
    j = i + 1
    while count != 0:
        if exp[j] == '(':
            count += 1
        if exp[j] == ')':
            count -= 1
        j += 1
        
    return evaluate(exp[i+1:j-1]), j

def evaluate(exp):    
    value = None
    operator = None
    
    i = 0
    while i < len(exp):
        if exp[i] in ['+', '*']:
            operator = exp[i]
            i += 1
            continue
        elif exp[i] == '(':
            v, i = eval_paren(exp, i)
        else:
            #print('hi', exp, i, exp[i])
            v = int(exp[i])
            i += 1
            
        if value is None:
            value = v
        else:
            if operator == '+':
                value += v
            elif operator == '*':
                value *= v
    
    return value
    
sum(evaluate(re.sub(r'\s', '', line)) for line in lines)
```

```python
# Part 2

def prod(iterable):
    return reduce(operator.mul, iterable, 1)

def eval_paren(exp, i):
    count = 1
    j = i + 1
    while count != 0:
        if exp[j] == '(':
            count += 1
        if exp[j] == ')':
            count -= 1
        j += 1
        
    return evaluate(exp[i+1:j-1]), j

def evaluate(exp):
    r = []
    
    i = 0
    while i < len(exp):
        if exp[i] in ['+', '*']:
            r.append(exp[i])
            i += 1
            continue
        elif exp[i] == '(':
            v, i = eval_paren(exp, i)
            r.append(v)
        else:
            r.append(int(exp[i]))
            i += 1
    
    for i in range(len(r)):
        if r[i] == '+':
            v = r[i-1] + r[i+1]
            r[i-1] = None
            r[i] = None
            r[i+1] = v
            i += 2
        else:
            i += 1
    
    r = [v for v in r if v]   
    return prod(v for v in r if v != '*')
            
    
sum(evaluate(re.sub(r'\s', '', line)) for line in lines)
```
