```python
import numpy as np
import re
```

```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
grid = np.zeros((1000, 1000), dtype=bool)

for l in lines:
    cmd, x1, y1, x2, y2 = re.match(r'(turn off|toggle|turn on) (\d+),(\d+) through (\d+),(\d+)\n', l).groups()
    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
    
    if cmd == 'turn off':
        grid[x1:x2 + 1, y1:y2+1] = False
    elif cmd == 'turn on':
        grid[x1:x2 + 1, y1:y2+1] = True
    else:
        grid[x1:x2 + 1, y1:y2+1] = ~grid[x1:x2 + 1, y1:y2+1]
        
grid.sum()
```

```python
grid = np.zeros((1000, 1000), dtype=int)

for l in lines:
    cmd, x1, y1, x2, y2 = re.match(r'(turn off|toggle|turn on) (\d+),(\d+) through (\d+),(\d+)\n', l).groups()
    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
    
    if cmd == 'turn off':
        grid[x1:x2 + 1, y1:y2+1] = np.maximum(grid[x1:x2 + 1, y1:y2+1] - 1, 0)
    elif cmd == 'turn on':
        grid[x1:x2 + 1, y1:y2+1] = grid[x1:x2 + 1, y1:y2+1] + 1
    else:
        grid[x1:x2 + 1, y1:y2+1] = grid[x1:x2 + 1, y1:y2+1] + 2
        
grid.sum()

```
