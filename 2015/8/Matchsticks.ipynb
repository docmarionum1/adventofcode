```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
# Part 1

char_count = 0
mem_count = 0

for l in lines:
    char_count += len(l.strip())
    mem_count += len(eval(l.strip()))
    
char_count - mem_count
```

```python
# Part 2

encoded_count = 0

for l in lines:
    encoded_count += len(l.strip()) + len([c for c in l.strip() if c in ['"', '\\']]) + 2
    
encoded_count - char_count
```
