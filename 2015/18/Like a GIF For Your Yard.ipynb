```python
import numpy as np
```

```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
# Part 1

grid = np.array([[True if c == '#' else False for c in l.strip()] for l in lines])

def next_state(grid, i, j):
    subgrid = grid[
        max(0, i-1):i+2, max(0, j-1): j+2
    ]
    
    num_on = subgrid.sum() - grid[i,j]
    
    if grid[i,j]:
        return num_on in [2, 3]
    else:
        return num_on == 3

for n in range(100):
    new_grid = np.copy(grid)
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            new_grid[i,j] = next_state(grid, i, j)
            
    grid = new_grid
    
grid.sum()
```

```python
# Part 2

grid = np.array([[True if c == '#' else False for c in l.strip()] for l in lines])
corner = len(grid) - 1
corners = [(0, 0), (0, corner), (corner, 0), (corner, corner)]
for i, j in corners:
    grid[i,j] = True

def next_state(grid, i, j):
    if (i, j) in corners:
        return True
    
    subgrid = grid[
        max(0, i-1):i+2, max(0, j-1): j+2
    ]
    
    num_on = subgrid.sum() - grid[i,j]
    
    if grid[i,j]:
        return num_on in [2, 3]
    else:
        return num_on == 3

for n in range(100):
    new_grid = np.copy(grid)
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            new_grid[i,j] = next_state(grid, i, j)
            
    grid = new_grid
    
grid.sum()
```
