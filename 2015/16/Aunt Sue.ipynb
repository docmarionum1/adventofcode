```python
import re
```

```python
with open('input.txt') as f:
    lines = f.readlines()
lines
```

```python
target = {
    'children': 3,
    'cats': 7,
    'samoyeds': 2,
    'pomeranians': 3,
    'akitas': 0,
    'vizslas': 0,
    'goldfish': 5,
    'trees': 3,
    'cars': 2,
    'perfumes': 1
}
```

```python
# Part 1

for l in lines:
    match = True
    num, compounds = re.match(r'Sue (\d+): (.*)\n', l).groups()
    for compound in compounds.split(','):
        name, number = compound.strip().split(':')
        if name in target and target[name] != int(number):
            match = False
            break
            
    if match:
        print(l)
```

```python
# Part 2

for l in lines:
    match = True
    num, compounds = re.match(r'Sue (\d+): (.*)\n', l).groups()
    for compound in compounds.split(','):
        name, number = compound.strip().split(':')
        if name in ['cats', 'trees']:
            if int(number) <= target[name]:
                match = False
                break
        elif name in ['pomeranians', 'goldfish']:
            if int(number) >= target[name]:
                match = False
                break
        elif name in target and target[name] != int(number):
            match = False
            break
            
    if match:
        print(l)
```
