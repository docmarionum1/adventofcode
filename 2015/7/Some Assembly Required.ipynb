```python
from lark import Lark, Transformer, v_args

wires = {}

l = Lark('''
start: statement "->" output

output: NAME
wire: NAME
value: wire
    | NUMBER

statement: value
    | value "AND" value -> and_
    | value "OR" value -> or_
    | wire "LSHIFT" value -> lshift
    | wire "RSHIFT" value -> rshift
    | "NOT" wire -> not_

%import common.CNAME -> NAME
%import common.NUMBER

%ignore " "
%ignore "\\n"
''')


for line in lines:
    tree = l.parse(line)
    
    wires[str(tree.children[1].children[0])] = tree.children[0]
```

```python
@v_args(inline=True)
class Eval(Transformer):
    def value(self, i):
        return int(i)
    
    def and_(self, x, y):
        return x & y
    
    def or_(self, x, y):
        return x | y
    
    def lshift(self, x, n):
        return x << n
    
    def rshift(self, x, n):
        return x >> n
    
    def not_(self, x):
        return x ^ 0xffff
    
    def wire(self, wire):
        #print(wire)
        if wire not in memo:
            memo[wire] = Eval().transform(wires[wire])
        
        return memo[wire]
         
    
    def statement(self, s):
        return s
```
